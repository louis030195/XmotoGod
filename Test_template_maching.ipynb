```python
from gym_xmoto.envs.capturedata import capturedata
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
SCREEN_HEIGHT, SCREEN_WIDTH  = 720, 480
```

```python
state = capturedata((80, 120, SCREEN_HEIGHT, SCREEN_WIDTH))
```

```python
state.shape
```

```python
template = cv2.imread('screenshots/win.png', 0)
w, h = template.shape[::-1]
template = cv2.resize(template , dsize=(int(w / 4), int(h / 4)))
gray_state = cv2.cvtColor(state, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray_state)
```

```python
res = cv2.matchTemplate(gray_state, template, cv2.TM_CCOEFF_NORMED)
threshold = 0.4
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(gray_state, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
plt.imshow(gray_state)
```
