```python
from gym_xmoto.envs.capturedata import capturedata
from matplotlib import pyplot as plt
import cv2
import numpy as np
```

```python
SCREEN_HEIGHT, SCREEN_WIDTH  = 720, 480
```

```python
state = capturedata((100, 120, SCREEN_HEIGHT, SCREEN_WIDTH))
```

```python
template = cv2.imread('gym_xmoto/envs/screenshots/skip_this_report.png', 0)
w, h = template.shape[::-1]
gray_state = cv2.cvtColor(state[1], cv2.COLOR_BGR2GRAY)
```

```python
print(gray_state.shape)
print(template.shape)
```

```python
plt.imshow(template)
```

```python
plt.imshow(gray_state)
```

```python
res = cv2.matchTemplate(gray_state, template, cv2.TM_CCOEFF_NORMED)
threshold = 0.7
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv2.rectangle(gray_state, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
plt.imshow(gray_state)
```
